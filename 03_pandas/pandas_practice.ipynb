**Inner Join**

An Inner join returns only the rows that have matching values in both tables. The resulting table contains only the rows where the join condition is satisfied. 

```python
pd.merge(left,right, how='inner', on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left,right,on=['key1', 'key2'])
```

**Outer Join**

An Outer join returns all the rows from both tables, and if there is no match in one of the tables, NULL values are returned.

```python
pd.merge(left,right,how= 'outer', on=['key1', 'key2'])
```

**Left Join**

A Left join returns all the rows from the left table (table1), and the matched rows from the right table (table2). If there is no match in the right table, NULL values are returned.

```python
pd.merge(left,right,how= 'left', on=['key1', 'key2'])
```

**Right Join**

A Right join is similar to a Left join, but it returns all the rows from the right table (table2) and the matched rows from the left table (table1). If there is no match in the left table, NULL values are returned. 

```python
pd.merge(left,right,how= 'right', on=['key1', 'key2'])
```

**Join vs. Merge**

In Merge, we combine two dataframes based on their columns, but in Join, we combine them based on their index.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3', 'C4'],
                    'D': ['D0', 'D2', 'D3', 'D4']},
                      index=['K0', 'K2', 'K3', 'k4'])
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
